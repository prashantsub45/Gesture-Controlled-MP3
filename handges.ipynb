```python
import cvzone
import pygame
from cvzone.HandTrackingModule import HandDetector
import cv2 as cv
```

```python
# class HandDetector:
    # ... (Your HandDetector class implementation here)

# Initialize Pygame
pygame.init()

# Load your music
pygame.mixer.music.load(r'C:\Users\ASUS\Desktop\opencv\BOLNAAA.mp3')
pygame.mixer.music.set_volume(0.5)  # Adjust the volume if needed

cap = cv.VideoCapture(0)
detector=HandDetector()

while True:
    success, img = cap.read()
    hands, img = detector.findHands(img)
    
    if hands:
         hands1 = hands[0]
         lmlist1 = hands1['lmList']
         bbox1 = hands1["bbox"]

         if lmlist1:
                fingers = detector.fingersUp(hands1)

                if fingers == [1, 1, 1, 1, 1]:
                    print('Play the music')
                if not pygame.mixer.music.get_busy():
                    pygame.mixer.music.play(-1)  # -1 makes the music loop indefinitely

                if fingers == [0, 0, 0, 0, 0]:
                    print('Pause the music')
                    pygame.mixer.music.pause()
```
